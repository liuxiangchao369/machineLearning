```python
import pandas as pd
from scipy import stats
import numpy as np
```

```python
raw_data = pd.read_csv("C:/Users/Nicolas/Documents/Scraping/thecarconnection/raw_data_cleaning_all_may.csv", index_col=0)
imputed_data = raw_data.copy()
```

```python
specs_to_mean = ['Front Shoulder Room (in)',
       'Front Head Room (in)', 'Second Leg Room (in)', 'Front Leg Room (in)',
       'Second Shoulder Room (in)', 'Second Head Room (in)',
       'Height, Overall (in)', 'Wheelbase (in)', 'Width, Max w/o mirrors (in)',
       'Third Gear Ratio (:1)', 'First Gear Ratio (:1)',
       'Fourth Gear Ratio (:1)', 'Second Gear Ratio (:1)',
       'Turning Diameter - Curb to Curb']
```

```python
imputed_data[specs_to_mean]
```

```python
list_col = imputed_data.columns.tolist()

specs_to_mode = list_col.copy()
```

```python
for item in specs_to_mean:
    specs_to_mode.remove(item)
    

specs_to_mode.remove('MSRP')
specs_to_mode.remove('Make')
specs_to_mode.remove('Model')
specs_to_mode.remove('Style Name')
specs_to_mode.remove('Body Style')
specs_to_mode.remove('EPA Classification')
```

```python
#specs_to_mode
```

```python
def imputer_mean(x, y):
    temp_list_model = raw_data.loc[raw_data["Model"] == raw_data.loc[x, "Model"]].loc[:, y].dropna()
    if len(temp_list_model) > 0:
        imputed_data.loc[x, y] = np.round(temp_list_model.mean(), 2)
    else:
        temp_list_epa = raw_data.loc[raw_data["EPA Classification"]
                                                == raw_data.loc[x, "EPA Classification"]].loc[:, y].dropna()
        if len(temp_list_epa) > 0:
            imputed_data.loc[x, y] = np.round(temp_list_epa.mean(), 2)
        else:
            temp_list_bodystyle = raw_data.loc[raw_data["Body Style"] ==
                                                        raw_data.loc[x, "Body Style"]].loc[:, y].dropna()
            if len(temp_list_bodystyle) > 0:
                imputed_data.loc[x, y] = np.round(temp_list_bodystyle.mean(), 2)
```

```python
def imputer_mode(x, y):
    temp_list_model = pd.Series(raw_data.loc[raw_data["Model"] == raw_data.loc[x, "Model"]].loc[:, y].dropna())
    if len(temp_list_model) > 0:
        imputed_data.loc[x, y] = stats.mode(temp_list_model)[0][0]
    else:
        temp_list_epa = raw_data.loc[raw_data["EPA Classification"]
                                                == raw_data.loc[x, "EPA Classification"]].loc[:, y].dropna()
        if len(temp_list_epa) > 0:
            imputed_data.loc[x, y] = stats.mode(temp_list_epa)[0][0]
        else:
            temp_list_bodystyle = raw_data.loc[raw_data["Body Style"] ==
                                                        raw_data.loc[x, "Body Style"]].loc[:, y].dropna()
            if len(temp_list_bodystyle) > 0:
                imputed_data.loc[x, y] = stats.mode(temp_list_bodystyle)[0][0]
```

```python
raw_data.isnull().sum().sum()
```

```python
ix = 0
for col in specs_to_mean:
    for row in imputed_data.index:
        val = imputed_data.at[row, col]
        if pd.isnull(val):
            raw_data.at[row,col] = imputer_mean(row, col)
    ix += 1
    print(str(ix)+'/'+str(len(specs_to_mean)), col)
```

```python
imputed_data.isnull().sum().sum()
```

```python
ix = 0
for col in specs_to_mode:
    for row in imputed_data.index:
        val = imputed_data.at[row, col]
        if pd.isnull(val):
            raw_data.at[row, col] = imputer_mode(row, col)
    ix += 1
    print(str(ix)+'/'+str(len(specs_to_mode)), col)
```

```python
specs_to_dummies = ['Drivetrain', 'Body Style',  'EPA Classification', 'Fuel System',
       'Steering Type', 'Rear Wheel Material', 'Suspension Type - Front', 'Air Bag-Frontal-Driver',
       'Air Bag-Frontal-Passenger', 'Air Bag-Passenger Switch (On/Off)',
       'Air Bag-Side Body-Front', 'Air Bag-Side Body-Rear',
       'Air Bag-Side Head-Front', 'Air Bag-Side Head-Rear', 'Brakes-ABS',
       'Child Safety Rear Door Locks', 'Daytime Running Lights',
       'Traction Control', 'Night Vision', 'Rollover Protection Bars',
       'Fog Lamps', 'Parking Aid', 'Tire Pressure Monitor', 'Back-Up Camera',
       'Stability Control', 'Hybrid Engine', 'Engine Configuration', 'Tire Rating', 
       'Country', 'Category']
```

```python
for item in specs_to_dummies:
    dummies = pd.get_dummies(imputed_data[item], prefix_sep=': ', prefix=item)
    imputed_data = pd.concat([imputed_data, dummies], sort=False, axis=1)

#df = df.drop(specs_to_dummies, axis=1)
imputed_data = imputed_data.drop('Other Features', axis=1)
```
